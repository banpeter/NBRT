```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from sklearn.model_selection import train_test_split
import csv
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=10000,
                n_features=4,
                centers=5,
                cluster_std=1,
                shuffle=True,
                random_state=2)



with open("DataSets/blob_att.csv", 'w',newline='') as f:
    # create the csv writer
    writer = csv.writer(f)

    # write a row to the csv file
    writer.writerow(["att1","att2","att3","att4"])
    #writer.writerow(["att1","att2"])
    for row in X:

        writer.writerow(row)


with open("DataSets/blob_label.csv", 'w',newline='') as f:
    # create the csv writer
    writer = csv.writer(f)

    # write a row to the csv file
    for row in y:
        writer.writerow([row])


        
plt.scatter(X[:,0],X[:,1],marker='o',c='blue',s=5)
plt.grid()
plt.show()
```

```python
#size: square meter math.sqrt(i+j/10)*7+ 0.2*i)*1000000
#distance from the center 1000/distance +10   dinstance 100-5000 meter
#government support yes = 0-10 million
#local average salaray(HUF) 2-8


X_training = []
y_training = []

header = ["size","distance","support","salary"]

for i in range(10000):
    
    price = 0
    
    size = random.randint(0,99)+random.randint(0,10)/10
    distance = random.randint(30,100)

    support = random.randint(0,10)
    salary = random.randint(3,8)

    price += ((math.sqrt(size)*10+0.3*(size/10)))
    price += (1000/distance + 5)

    price += support
    price += salary**3
    

    X_training.append([size,distance,support,salary])
    y_training.append([price])
    
    
print(len(X_training))
print(len(y_training))    
    


with open("DataSets/house_att.csv", 'w',newline='') as f:
    # create the csv writer
    writer = csv.writer(f)

    # write a row to the csv file
    writer.writerow(header)
    for row in X_training:

        writer.writerow(row)



with open("DataSets/house_price.csv", 'w',newline='') as f:
    # create the csv writer
    writer = csv.writer(f)

    # write a row to the csv file
    #writer.writerow(header)
    for row in y_training:

        writer.writerow(row)

```
