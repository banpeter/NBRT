```python
#Execution time
times = []

with open("DataSets/times_rt.csv","r") as file:
    t = []
    tmp = []
    for row in file:
        t.append(float(row.strip("\n")))
    for i in range(4):
        print(len(t))
        tmp.append((t[i]+t[i+4]+t[i+8]+t[i+12]+t[i+16])/5)
    times.append(tmp)
    #print(times)

   
    
with open("DataSets/times_rtl.csv","r") as file:
    t = []
    tmp = []
    for row in file:
        t.append(float(row.strip("\n")))
    for i in range(4):
        #print(len(t))
        tmp.append((t[i]+t[i+4]+t[i+8]+t[i+12]+t[i+16])/5)
    times.append(tmp)
    #print(times)
 
with open("DataSets/times_rtln.csv","r") as file:
    t = []
    tmp = []
    temp = []
    for row in file:
        #print(row)
        row = row.strip('\n').split(',')
        #print(row)
        t.append([float(x) for x in row ])

    for i in range(4):
        print(t[i])
        tmp = []
        for j in range(4):
            tmp.append( (t[i][j]+t[i+4][j]+t[i+8][j]+t[i+12][j]+t[i+16][j])/5   )
        temp.append(tmp)
    print(temp)
    #print(t)
    times.append(temp)

   
    
with open("DataSets/times_rtn.csv","r") as file:
    t = []
    tmp = []
    temp = []
    for row in file:
        #print(row)
        row = row.strip('\n').split(',')
        #print(row)
        t.append([float(x) for x in row ])

    for i in range(4):
        print(t[i])
        tmp = []
        for j in range(4):
            tmp.append( (t[i][j]+t[i+4][j]+t[i+8][j]+t[i+12][j]+t[i+16][j])/5   )
        temp.append(tmp)
    print(temp)
    #print(t)
    times.append(temp)
            
    

    
'''with open("DataSets/times_rtni.csv","r") as file:
    t = []
    
    for row in file:
        #print(row)
        row = row.strip('\n').split(',')
        #print(row)
        t.append([float(x) for x in row ])
    
    times.append(t)'''
    
```
