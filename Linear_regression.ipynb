```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("FuelConsumption.csv")
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]

train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
```

```python
def gradient(a,b,X,y,L):
    
    da = 0
    db = 0
    
    for i in range(len(X)):
        
        da += a*X[i]**2 + b*X[i] - X[i]*y[i]
        db += a*X[i] + b - y[i]
        #-x[i]**2 * y +x[i]*y[i]
        #y[i] * (x[i]**2 + x[i
        
    a = a - da*L
    b = b - db*L
    
   
    
    return a,b,0
    
    
    
```

```python
L = 0.0001
epochs = 1000

b = 0
m = 0

for i in range(epochs):
    
    m,b,t = gradient(m,b,train_x,train_y,L)
    if(t):
        break


plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.plot(cdf.ENGINESIZE, [m*x+b for x  in cdf.ENGINESIZE], c = "r")

plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
print(m)
print(b)
```
